### Metric Evaluation
---

Type 1 error    : False Positive  
Kondisi         : Model memprediksi customer membatalkan booking hotel namun aktualnya lanjut.
Konsekuensi     : Perusahaan akan membuang waktu dan biaya untuk menangani hal yang tidak perlu ditangani.

Type 2 error    : False Negative  
Kondisi         : Model memprediksi customer melanjutkan booking hotel namun aktualnya membatalkan.
Konsekuensi     : Perusahaan akan kehilangan potensial revenue.

Dalam konteks ini, dengan fokus tujuan yaitu menangani pelanggan yang terindikasi membatalkan, perusahaan perlu mengembangkan model yang dapat mengurangi Type 2 error (False Negative). Type 2 error terjadi ketika model memprediksi bahwa pelanggan akan melanjutkan booking hotel, tetapi sebenarnya mereka membatalkannya. Konsekuensinya adalah perusahaan akan kehilangan potensial pendapatan.

Untuk mengatasi masalah ini, perusahaan harus memprioritaskan `Recall` sebagai metrik utama. `Recall` mengukur kemampuan model dalam mengidentifikasi sebagian besar kasus positif yang sebenarnya. Dalam hal ini, recall akan mengukur seberapa baik model dapat mengenali pelanggan yang sebenarnya akan membatalkan. Dengan meningkatkan `Recall`, perusahaan dapat mengurangi jumlah False Negative, sehingga dapat mengidentifikasi lebih banyak pelanggan yang berpotensi membatalkan dan mengambil langkah-langkah yang diperlukan untuk mempertahankan mereka.

Pada bagian berikut akan dilakukan proses penanganan outlier yang mengganggu interpretasi analisa dan keakuratan model.

```python
# Distribusi Data ADR
plt.figure(figsize=(15,5))

plt.subplot(121)
sns.histplot(data=df,x='adr',kde=True)
plt.title('Average Daily Rate Histogram',fontsize=20)

plt.subplot(122)
sns.boxplot(data=df,y='adr')
plt.title('Average Daily Rate Boxplot',fontsize=20)
```

```python
# Menghilangkan data outlier dengan pendekatan IQR
Q1 = df['adr'].quantile(0.25)
Q3 = df['adr'].quantile(0.75)

IQR = Q3 - Q1

batas_atas = Q3 + 1.5 * IQR
batas_bawah = Q1 - 1.5 * IQR

df = df[(df['adr'] >= batas_bawah) & (df['adr'] <= batas_atas)]

df.shape
```

```python
# Distribusi Data ADR Setelah Outlier Handling
plt.figure(figsize=(15,5))

plt.subplot(121)
sns.histplot(data=df,x='adr',kde=True)
plt.title('Average Daily Rate Histogram',fontsize=20)

plt.subplot(122)
sns.boxplot(data=df,y='adr')
plt.title('Average Daily Rate Boxplot',fontsize=20)
```

Setelah memeriksa dataset, untuk kolom `adr` yang memiliki nilai average daily rate terlihat adanya nilai ekstrem dari range wajarnya sehingga dilakukan proses handling outlier. Pendekatan handling Outlier dilakukan dengan IQR dimana dengan metode ini akan ditangani data ekstrem yang berada jauh dari nilai-nilai data lainnya dalam sebuah dataset.

### Transforming & Encoding
Fitur-fitur akan mengalami proses encoding dan scaling berdasarkan jenis dan karakteristiknya. Pada dataset ini, semua fitur numerik akan di-scale menggunakan StandardScaler, dan semua fitur kategorikal akan di-encode menggunakan OneHotEncoder.

- RobustScaler: 
    RobustScaler dipilih sebagai metode scaling untuk fitur `lead_time`, `arrival_date_year`, `arrival_date_week_number`, `arrival_date_day_of_month`, `stays_in_weekend_nights`, `stays_in_week_nights`, `adults`, `children`, `babies`, `is_repeated_guest`, `previous_cancellations`, `previous_bookings_not_canceled`, `booking_changes`, `agent`, `days_in_waiting_list`, `adr`, `required_car_parking_spaces`, `total_of_special_requests`, `reserved_room_type`, `assigned_room_type`, dan `arrival_date_month`. RobustScaler sangat berguna ketika menghadapi fitur-fitur numerik yang mengandung outlier atau tidak terdistribusi secara normal. Skala RobustScaler menggunakan statistik yang kuat terhadap nilai ekstrem. Dengan mengurangkan median dan membagi dengan jangkauan interkuartil, RobustScaler melakukan scaling pada fitur-fitur dengan cara yang kurang dipengaruhi oleh outlier, sehingga memastikan transformasi yang lebih andal dan akurat.

- OneHotEncoder:

    OneHotEncoder digunakan untuk fitur `hotel`, `meal`, `market_segment`, `distribution_channel`, `deposit_type`, dan `customer_type`. OneHotEncoder cocok untuk fitur-fitur kategorikal dengan jumlah nilai unik yang kecil dan tanpa urutan inheren. Ini menciptakan kolom biner untuk setiap nilai unik, menunjukkan keberadaan atau ketiadaan nilai tersebut dalam fitur asli. Skema encoding ini memungkinkan model pembelajaran mesin untuk dengan efektif menafsirkan dan menggunakan data kategorikal, menangkap kategori-kategori yang berbeda tanpa memberlakukan urutan numerik apapun.

- BinaryEncoder:

    BinaryEncoder digunakan untuk mengubah variabel kategorikal `country` menjadi representasi biner. Metode encoding ini memungkinkan kita untuk mengatasi data kategorikal dengan efisien dengan mewakili setiap kategori unik dengan pola bit biner. Ini memungkinkan kita untuk menyimpan informasi kategorikal secara ringkas dan menggunakannya dalam berbagai analisis atau algoritma pembelajaran mesin.

### Benchmarking
Pada bagian ini, akan dilakukan benchmarking menggunakan berbagai variasi algoritma dan perlakuan untuk mengatasi ketidakseimbangan data, serta menggunakan teknik cross-validation. Tujuan dari benchmarking ini adalah untuk memperoleh estimasi seberapa baik model dapat di andalkan akan terhadap data yang belum dilihat sebelumnya.

```python
# Base Model (Basic ML algorithm)
logreg = LogisticRegression(random_state=42,class_weight='balanced')
knn = KNeighborsClassifier()
tree = DecisionTreeClassifier(random_state=42,class_weight='balanced')

# Modern Algorithm
xgb = XGBClassifier(random_state = 42)
lgbm = LGBMClassifier(random_state = 42,class_weight='balanced')
cat = CatBoostClassifier(random_state=42, verbose=False)
rf = RandomForestClassifier(random_state=42,class_weight='balanced')
gboost = GradientBoostingClassifier(random_state=42)
bag = BaggingClassifier(random_state=42)
ada = AdaBoostClassifier(random_state=42)
```

Analisis ini akan melibatkan berbagai model klasifikasi,diantaranya base model diantaranya Logistic Regresion, K-Nearest Neighbors, Decision Tree. Selain itu, algoritma modern seperti XGBoost, LightGBM, CatBoost,Random Forest, Gradient Boosting, Bagging, dan AdaBoost akan digunakan untuk meningkatkan performa model. Parameter class_weight='balanced' akan digunakan pada model Logistic Reresion, Decision Tree, Light Gradient Boosting dan Random Forest untuk mengatasi distribusi kelas yang tidak seimbang.

```python
f2_score = make_scorer(fbeta_score, beta=2)

ros = RandomOverSampler(random_state=42)
rus = RandomUnderSampler(random_state=42)
smote = SMOTE(random_state=42)
near = NearMiss()

# Define imbalance treatment methods
models = [logreg, knn, tree, xgb, lgbm, cat, rf, gboost, bag, ada]

resample = {
    'none': None,
    'ros': ros,
    'rus': rus,
    'smote': smote,
    'near': near,
}

metrics = {'balanced_accuracy': 'balanced_accuracy',
           'precision': 'precision',
           'recall': 'recall',
           'f1': 'f1',
           'f2': f2_score}

fit_time_arr = []
score_time_arr = []     
test_accuracy_arr = []    
test_precision_arr = []
test_recall_arr = []
test_f1_arr = []
test_f2_arr = []


for key,val in resample.items():
    for i in models:
        crossval = StratifiedKFold(n_splits=5)

        if val == None:
            estimator = Pipeline([
                ('transorm', transformer),
                ('model', i),
            ])
        else:
            estimator = Pipeline([
                ('transorm', transformer),
                ('resample', val),
                ('model', i),
            ])

        scores = cross_validate(estimator, X_train, y_train,cv=crossval, scoring=metrics)
        fit_time_arr.append(scores['fit_time'].mean())
        score_time_arr.append(scores['score_time'].mean())
        test_accuracy_arr.append(scores['test_balanced_accuracy'].mean())
        test_precision_arr.append(scores['test_precision'].mean())
        test_recall_arr.append(scores['test_recall'].mean())
        test_f1_arr.append(scores['test_f1'].mean())
        test_f2_arr.append(scores['test_f2'].mean())

list_algo = []
list_resample = []


for key in resample.keys():
    for mod in ['logreg','knn','tree', 'xgb', 'lgbm','cat','rf','gboost','bag','ada']:
        list_algo.append(mod)
        list_resample.append(key)

model_bench = pd.DataFrame({
    'model': list_algo,
    'resample': list_resample,
    'fit_time': fit_time_arr,
    'score_time': score_time_arr,
    'accuracy': test_accuracy_arr,
    'precision': test_precision_arr,
    'recall': test_recall_arr,
    'f1': test_f1_arr,
    'f2': test_f2_arr
})
```

#### CatBoost (Category Boost) 

Pada bechmarking sebelumnya CatBoost menjadi Alogritma yang terbaik berdasarkan matrix recall, algoritma ini sendiri merupakan sebuah algoritma pembelajaran mesin yang dirancang khusus untuk menangani fitur kategori (categorical features) dengan efisien. Algoritma ini merupakan perluasan dari kerangka kerja gradient boosting dan memiliki kesamaan dengan algoritma seperti XGBoost dan LightGBM.CatBoost menonjol karena kemampuannya untuk secara otomatis mengatasi variabel kategorikal tanpa memerlukan encoding eksplisit, sehingga menjadi alat berharga untuk dataset dunia nyata dengan tipe data yang beragam. 

CatBoost menggunakan algoritma canggih yang disebut Ordered Boosting, yang menggabungkan gradient boosting dan statistik target yang diurutkan.
Teknik ini mengatasi tantangan dalam bekerja dengan fitur-fitur kategorikal dengan memperlakukannya sebagai variabel numerik, memberikan representasi yang lebih akurat tentang pola dan ketergantungan yang mendasari.Salah satu keunggulan utama dari CatBoost adalah kemampuannya untuk menangani variabel kategorikal dengan berbagai kardinalitas dan tingkat.

Secara internal, CatBoost memproses fitur-fitur kategorikal dengan mempertimbangkan kombinasi kategori dan interaksi di antara mereka, sehingga membantu menangkap informasi berharga yang mungkin terlewat oleh algoritma lain.

Beberapa fitur menonjol dari CatBoost meliputi:

1. Penanganan variabel kategorikal 
    CatBoost mengatasi fitur-fitur kategorikal secara mulus, menghilangkan kebutuhan untuk encoding eksplisit atau teknik feature engineering.

2. Penanganan otomatis nilai yang hilang 
    CatBoost memiliki kemampuan bawaan untuk mengatasi nilai yang hilang baik dalam fitur numerik maupun kategorikal selama pelatihan, menyederhanakan langkah pra-pemrosesan.

3. Ketahanan terhadap pencilan (outliers)
    CatBoost dirancang agar tahan terhadap data pencilan, mengurangi dampaknya pada prediksi model.

4. Pelatihan dan inferensi yang cepat 
    CatBoost memanfaatkan teknik paralelisasi untuk mempercepat proses pelatihan, sehingga cocok untuk dataset besar.

5. Interpretabilitas 
    CatBoost memberikan wawasan tentang pentingnya fitur, memungkinkan pengguna untuk memahami fitur mana yang berkontribusi paling besar terhadap prediksi model.

CatBoost telah berhasil diterapkan dalam berbagai domain, termasuk e-commerce, periklanan, dan sistem rekomendasi.Kemampuannya untuk mengatasi fitur kategorikal dengan efektif dan kinerjanya yang kompetitif membuatnya menjadi alat yang berharga dalam toolkit data scientist, terutama ketika berurusan dengan dataset dunia nyata yang seringkali mengandung campuran variabel numerik dan kategorikal.

## Hyperparameter Tuning

Selanjutnya, fokus akan berada pada mengoptimalkan kinerja model CatBoost melalui penyetelan hyperparameter. Dengan menyelaraskan kembali hyperparameter model, tujuannya adalah untuk meningkatkan efektivitasnya dan mencapai hasil yang lebih baik lagi.

### Confussion Matrix Before Tuning

```python
y_pred = model.predict(X_test)
cf_matrix = confusion_matrix(y_test, y_pred, labels=model.classes_)
group_names = ['True Neg', 'False Pos', 'False Neg', 'True Pos']
group_counts = ['{0:0.0f}'.format(value) for value in cf_matrix.flatten()]
group_percentages = ['{0:.2%}'.format(value) for value in cf_matrix.flatten() / np.sum(cf_matrix)]
labels = [f'{v1}\n{v2}\n{v3}' for v1, v2, v3 in zip(group_names, group_counts, group_percentages)]
labels = np.asarray(labels).reshape(2, 2)
sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues')
plt.xlabel('Predict Label Before')
plt.ylabel('True Label Before')
plt.title('Confusion Matrix Before tuning')
```

```python
# Di bagian bawah ini langsung ditampilkan parameter terbaik hasil tuning karena pertimbangan terlalu lama saat running code nya.

# Hyperparameter space untuk CatBoost
# hyperparam_space = {
#     'model__depth': [i for i in range(1, 11)],
#     'model__iterations': [100, 150, 200, 250, 300],
#     'model__learning_rate': [0.1, 0.01, 0.001, 0.0001],
#     'model__l2_leaf_reg': [1, 3, 5, 7],
#     'model__bagging_temperature': [0.8, 1, 1.2, 1.5],
# }

hyperparam_space = {
    'model__depth': [10],
    'model__iterations': [350],
    'model__learning_rate': [0.1],
    'model__bagging_temperature': [1.2],
}
```

Berikut adalah penjelasan mengenai setiap hyperparameter pada model:

1. `model__depth`: Ini adalah hyperparameter yang menentukan kedalaman maksimum dari decision trees dalam model CatBoost. Kedalaman pohon mempengaruhi kompleksitas model dan kemampuannya untuk menangkap hubungan yang lebih kompleks dalam data. Semakin dalam pohon, semakin kompleks modelnya, namun ini juga bisa meningkatkan risiko overfitting. Nilai yang lebih tinggi dari `model__depth` akan memperbolehkan model untuk membuat aturan yang lebih kompleks dan detail.

2. `model__iterations`: Ini adalah hyperparameter yang menentukan jumlah maksimum iterasi (putaran) yang akan dilakukan oleh model CatBoost. Setiap iterasi menambahkan satu decision tree ke dalam model. Semakin banyak iterasi, semakin banyak decision tree yang akan digunakan dalam model, dan ini dapat meningkatkan akurasi model. Namun, perlu diingat bahwa nilai yang terlalu tinggi dari `model__iterations` juga bisa menyebabkan overfitting.

3. `model__learning_rate`: Ini adalah hyperparameter yang menentukan tingkat pembelajaran dalam model CatBoost. Tingkat pembelajaran mengendalikan seberapa cepat model belajar dari kesalahan prediksi sebelumnya. Nilai yang lebih tinggi dari `model__learning_rate` cenderung mengarah pada konvergensi yang lebih cepat, namun juga meningkatkan risiko overfitting. Tingkat pembelajaran yang tepat dapat membantu mencapai keseimbangan antara konvergensi yang cepat dan akurasi yang baik.

4. `model__bagging_temperature`: Ini adalah hyperparameter yang mengatur tingkat perubahan suhu dalam metode bagging pada model CatBoost. Bagging adalah teknik penggabungan (ensembling) beberapa model untuk meningkatkan akurasi. Dengan mengatur `model__bagging_temperature`, kita bisa mengendalikan seberapa acak model akan dipilih untuk penggabungan. Nilai yang lebih rendah akan membuat model lebih acak, sementara nilai yang lebih tinggi akan membuat model lebih deterministik.

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV,KFold

crossval_x = KFold(n_splits=5, shuffle=True, random_state=42)

f2_scorer = make_scorer(fbeta_score, beta=2)

pipe_model = Pipeline([
('prep', transformer),
('resampling', rus),
('model', cat)
])

grid_search = GridSearchCV(
    estimator=pipe_model,
    param_grid=hyperparam_space,
    cv=crossval_x,
    scoring='recall',
    n_jobs=-1,
)
```
